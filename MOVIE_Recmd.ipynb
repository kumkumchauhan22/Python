```python
# main.py

movies = [
    {"title": "Inception", "genre": "Sci-Fi"},
    {"title": "The Matrix", "genre": "Sci-Fi"},
    {"title": "The Godfather", "genre": "Crime"},
    {"title": "Pulp Fiction", "genre": "Crime"},
    {"title": "The Dark Knight", "genre": "Action"},
    {"title": "Avengers: Endgame", "genre": "Action"},
    {"title": "Forrest Gump", "genre": "Drama"},
    {"title": "The Shawshank Redemption", "genre": "Drama"}
]

def recommend_by_genre(genre):
    recommendations = [movie["title"] for movie in movies if movie["genre"].lower() == genre.lower()]
    return recommendations

def main():
    print("=== Movie Recommendation System ===")
    genre = input("Enter a genre (e.g., Sci-Fi, Crime, Action, Drama): ")

    results = recommend_by_genre(genre)

    if results:
        print(f"\nRecommended {genre.title()} Movies:")
        for movie in results:
            print(f"- {movie}")
    else:
        print("No movies found for that genre.")

if __name__ == "__main__":
    main()
```
